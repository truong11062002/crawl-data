```python
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
def rename_crawl(img_folder, csv_file, prefix_name):
    annot = pd.read_csv(csv_file)
    folder_name = os.listdir(img_folder)
    name_csv = annot['name']
    print('Đổi tên cho ảnh crawl')
    for i in tqdm(range(len(name_csv))):
        if name_csv[i] in folder_name:
            src = os.path.join(img_folder, name_csv[i])
            dst = prefix_name + str(i).zfill(5) + ".jpg"
            name_csv[i] = dst
            dst = os.path.join(img_folder, name_csv[i])
            os.rename(src, dst)
    print('Cập nhật lại csv')
    annot = pd.concat([pd.DataFrame(np.array(annot['url'])), 
                       pd.DataFrame(name_csv)], axis=1)
    annot.columns = ['url', 'name']
    annot.to_csv(csv_file)
```

```python
def rename_collect(img_folder, csv_file, prefix_name):
    folder_name = os.listdir(img_folder)
    annot = pd.read_csv(csv_file)
    folder_name = os.listdir(img_folder)
    name_csv = annot['name']
    print('Đổi tên cho ảnh tụ thu thập')
    dem=0
    for i in tqdm(range(len(folder_name))):
        if not folder_name[i].startswith(prefix_name):
            dem+=1
            src = os.path.join(img_folder, folder_name[i])
            dst = prefix_name + str(len(name_csv) + dem).zfill(5) + ".jpg"
            dst = os.path.join(img_folder, dst)
            os.rename(src, dst)    
```

```python
prefix_name = 'mut_quat' 
rename_crawl('/content/drive/MyDrive/DoAn_BanhMut/MUT QUAT', '/content/drive/MyDrive/DoAn_BanhMut/muc_quat.csv', prefix_name)
rename_collect('/content/drive/MyDrive/DoAn_BanhMut/MUT QUAT','/content/drive/MyDrive/DoAn_BanhMut/muc_quat.csv',prefix_name)
```
