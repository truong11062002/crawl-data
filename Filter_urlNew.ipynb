```
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```
def merge_url(file): 
    img_link = []
    img_name = []
    temp_file = pd.read_csv(file, header = None, error_bad_lines = False, warn_bad_lines = False)
    for link in tqdm(temp_file.iloc[:,0]):
        img_link.append(link)
        img_name.append(link.rsplit('/', 1)[-1])
    url = pd.concat([pd.DataFrame(img_link), pd.DataFrame(img_name)], axis = 1)
    url.columns = ['url', 'name']
    return url
```

```
def remove(imgs_folder, url_file, save_url):
    img_names = os.listdir(imgs_folder)
    urls = np.array(url_file.iloc[:,1].values)
    check = []
    for temp in urls:
        if temp in img_names:
            check.append(temp)
    check = np.unique(check)
    val = list(url_file.iloc[:,0].values)
    key = list(url_file.iloc[:,1].values)
    dict_url = dict(zip(key, val))
    filted_url = []
    for i in check:
        filted_url.append(dict_url[i])
    filted_url = pd.concat([pd.DataFrame(filted_url), pd.DataFrame(check)], axis  = 1)
    filted_url.columns = ['url', 'name']
    filted_url.to_csv(save_url, index = None)
    return filted_url
```
