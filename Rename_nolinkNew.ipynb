```python
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
def rename_nolink(img_folder, prefix_name):
    folder_name = os.listdir(img_folder)
    print('Đổi tên cho ảnh tự thu thập')
    for i in tqdm(range(len(folder_name))):
            src = os.path.join(img_folder, folder_name[i])
            dst = prefix_name + str(i).zfill(5) + ".jpg"
            dst = os.path.join(img_folder, dst)
            os.rename(src, dst)    
```

```python
prefix_name = 'mut_quat_' 
rename_nolink('/content/drive/MyDrive/DoAn_BanhMut/MUT QUAT',prefix_name)
```
